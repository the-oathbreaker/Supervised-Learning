```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
import sklearn
import pandas as pd
```

```python
df=pd.read_excel('E:\ml\ptry.xls')
```

```python
df_x=df.iloc[:,0:13]
df_y=df.iloc[:,13]
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=4)
x_train
```

```python
alpha=0.01
j_history=np.zeros((2500,1))
```

```python
x_train=x_train.as_matrix()
y_train=y_train.as_matrix()
```

```python
def scaling(x):
    a=np.sum(x)/len(x)
    s=np.std(x)
    for i in range(len(x)):
        x[i]=(x[i]-a)/s
    return x
```

```python
for i in range(13):
    x_train[:,i]=scaling(x_train[:,i])
    
```

```python
y_train=y_train.reshape((y_train.shape[0],1))
```

```python
x1=np.c_[np.ones((404,1)),x_train]
```

```python
t=np.zeros((14,1))
```

```python
def cost(x,y,t):
    q=np.dot(x,t)-y
    j=(np.sum(q*q))/(2*len(x))
    return j
```

```python
for i in range(2500):
    q=np.dot(x1,t)-y_train
    t=t-np.dot(x1.T,q)*(alpha/len(x1))
    j=cost(x1,y_train,t)
    j_history[i]=j
```

```python
p=np.arange(1,2501,1)
plt.plot(p,j_history)
plt.show()
```

```python
x2=np.c_[np.ones((len(x_test),1)),x_test]
```

```python
y_test=y_test.reshape((len(y_test),1))
```

```python
for i in range(1,14):
    x2[:,i]=scaling(x2[:,i])
```

```python
x_predict=np.dot(x2,t)
```

```python
v=np.arange(61)
plt.plot(x_predict,y_test,'or')
plt.plot(v,v)
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
#accuracy
r2_score(y_test,x_predict)
```
